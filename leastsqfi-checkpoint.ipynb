```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Input: (t,y)=data set, n= degree of the polynomial model 
#Output:
def leastsqfit(t, y, n):
    m = t.size # number of data points
    d = n+1 # number of coefficients to be determined
    # ATA x=ATb is the normal equation to solve
    ATA = np.zeros((d, d))
    ATb = np.zeros(d)
    p = np.zeros(2*n+1)
    for k in range(d):
        sum = 0
        for i in range(m):
            sum =sum+ y[i]*t[i]**k
        ATb[k] = sum
    # p[i] below is the sum of the i-th power of the t coordinates
    p[0] = m
    for i in range(1, 2*n+1):
        sum = 0
        for j in range(m):
            sum = sum+ t[j]**i
        p[i] = sum
    # Compute the upper triangular part of the coefficient
    # matrix ATA, and its diagonal
    for k in range(d):
        for j in range(k, d):
            ATA[k, j] = p[k+j]
            # The lower triangular part of the matrix is defined using the
            # fact the matrix ATA is symmetric
            for i in range(1, d):
                for j in range(i):
                    ATA[i, j] = ATA[j, i]
    x = np.linalg.solve(ATA,ATb)
    return(x)
```
