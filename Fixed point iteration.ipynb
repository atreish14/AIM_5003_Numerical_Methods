```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def fixedp(f,x0,TOL=10e-5,N=100):
#Fixed point algorithm 
    e = 1
    i = 0
    xi = []
    while(e > TOL and i < N):
        x=f(x0)      # fixed point equation
        e= np.abs(x0-x) # error at the current step
        x0=x
        xi.append(x0)  # save the solution of the current step
        i= i + 1
    return(xi,i)
    if e< TOL:
        print('Approximate fixed point', x0) 
    else:
        print('The fixed point iteration diverges') 
```

```python
def g1(x):
    return  1-x**3
```

```python
fixedp(g1,0.5,10e-5,100)
```

```python
def g2(x):
    return  (1-x)**(1/3)
```

```python
fixedp(g2,0.5,10e-5,100)
```

```python
def g3(x):
    return  r*(x-x**2)
```

```python
r=3.8
(xi,i)=fixedp(g3,0.5,10e-5,100)
```
