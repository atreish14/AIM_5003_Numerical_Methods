```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Euler method for solving differential equations
#Input: d=dimension of the system of equations
#       f=vector field
#       (t0,y0)=initial condition
#       t_final= final time
#       N=number of steps
#Output:t=time grid, y=approximate solution 
def Euler(d,f,t0,y0,t_final,N):
    t=np.zeros(N+1)
    y=np.zeros((N+1,d))
    h=(t_final-t0)/N
    #print(h)
    t[0]=t0
    y[0]=y0
    for i in range(N):
        t[i+1]=t[i]+h
        y[i+1]=y[i]+h*f(t[i],y[i])
    return t, y 
```

```python
def f(t,y):
    return t*y+t**3

def y_exact(t):
    return 3*np.exp(0.5*t**2)-t**2-2
```

```python
d=1;t0=0.;y0=1.;t_final=1.;N=10
t,y=Euler(d,f,t0,y0,t_final,N)
t_axis = np.linspace(t0, t_final, 100)
y_vals = y_exact(t_axis)
plt.plot(t_axis, y_vals)
plt.plot(t, y, 'o');
```

```python
#Harmonic oscillator (mass-spring system)
def F(t,y):
    return np.array([y[1],-y[0]])
```

```python
d=2;t0=0.;y0=np.array([1.,0]);t_final=6.28;N=100
t,y=Euler(d,F,t0,y0,t_final,N)
t_axis = np.linspace(t0, t_final, 100)
plt.plot(y[:,0],y[:,1], 'b-') #trajectory
plt.plot([y[0,0]], [y[0,1]], 'go') # start
plt.show()
plt.plot(t,y)
plt.show()
```
